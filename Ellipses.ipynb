```python
#Preamble
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generateEllipse(a,b,gridX,gridY):
    grid = np.zeros([gridX,gridY])
    centerX = int(gridX/2)
    centerY = int(gridY/2)
    for x in range(len(grid)):
        for y in range(len(grid[0])):
            if a > 0 and b > 0:
                if ((x-centerX)/a)*((x-centerX)/a) + ((y-centerY)/b)*((y-centerY)/b) <= 1:
                    grid[x][y] = 1
    
    return grid
```

```python
ellipse = generateEllipse(100,50,200,200)
area = np.sum(ellipse)
trueArea = np.pi*100*100
areaError = (area-trueArea)/trueArea
print(areaError)

fig, ax = plt.subplots()
ax.imshow(ellipse, cmap = 'Greys', aspect = 'equal', origin = 'upper')
plt.show()
```

```python
fig, axs = plt.subplots(5,5, figsize = (12,12))
for i in range(25):
    x = 10*(i+1)
    y = 10*(i+1)
    b = int(x/2)
    a = int(b/2)
    
    ellipse = generateEllipse(a,b,x,y)
    area = np.sum(ellipse)
    trueArea = np.pi*a*b
    areaError = (area-trueArea)/trueArea
    axs[int(i/5)][i%5].imshow(ellipse, cmap = 'Greys', aspect = 'equal', origin = 'upper')
    print(areaError)

plt.show()
```
